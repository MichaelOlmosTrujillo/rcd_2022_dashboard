## Analisis de calidad de datos del anexo VIII

### Valores nulos en los datos

```python
pd.set_option('display.max_rows', None)
df_nulos = (df.isnull().sum()/df.shape[0]
    ).to_frame().reset_index().rename(columns={'index':'columna',
                                                0:'cantidad_nulos'})
## Round cantidad_nulos to 2 decimal places

df_nulos['cantidad_nulos'] = df_nulos['cantidad_nulos'].round(2)
df_nulos = df_nulos.rename(columns={'cantidad_nulos':'porcentaje_nulos'})
df_nulos
```

Se descarta la columna Autoridad Ambiental del análisis de datos nulos dado que no tiene datos nulos

```python
df_nulos = df_nulos[df_nulos['columna'] != 'autoridad_ambiental']
df_nulos
```

### Promedio de datos nulos

```python
df_nulos['porcentaje_nulos'].mean()
```

```python
plt.figure().set_size_inches(8, 10, forward=True)
sns.barplot(data=df_nulos,y='columna',
            x='porcentaje_nulos',
            hue = 'porcentaje_nulos',
            dodge=False
            )
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.yticks(fontsize = 6)
plt.tight_layout()
plt.xlabel('Porcentaje de valores Nulos')
plt.ylabel('Columnas')
plt.title('Porcentaje de valores nulos por campo del Anexo VIII')
```

## Tipo de datos en cada Campo

```python
def find_unique_dtypes(df):
    dtype_info = []
    for column in df.columns:
        fila = {}
        unique_types = {type(x) for x in df[column]}
        cantidad_de_tipo_de_datos = len(unique_types)
        #fila[column] = unique_types
        fila['columna'] = column
        fila['tipos_de_datos'] = unique_types
        fila['cantidad_de_tipo_de_datos'] = cantidad_de_tipo_de_datos
        dtype_info.append(fila)
    return dtype_info
```

```python
df_tipos_de_datos = pd.DataFrame(find_unique_dtypes(df))
df_tipos_de_datos
```

```python
df['total_rcd_gen_obra_ton'][df['total_rcd_gen_obra_ton'].apply(lambda x: isinstance(x, str))].to_frame()
```

## Análisis de datos

### Cantidad en toneladas del total de RCD generados en obra

```python
rcd_gen_obra = df[['autoridad_ambiental', 
                   'total_rcd_gen_obra_ton']]
rcd_gen_obra
```

#### Quitando la coma de los valores

```python
# Function to remove commas from strings
def remove_commas(value):
    if isinstance(value, str):
        return float(value.replace(',', ''))
    return value
```

```python
rcd_gen_obra['total_rcd_gen_obra_ton'] = rcd_gen_obra['total_rcd_gen_obra_ton'].apply(remove_commas)
```

```python
rcd_gen_obra['total_rcd_gen_obra_ton'] = rcd_gen_obra['total_rcd_gen_obra_ton'].apply(lambda x: '{:.3f}'.format(x))
```

```python
rcd_gen_obra['total_rcd_gen_obra_ton'] = rcd_gen_obra['total_rcd_gen_obra_ton'].astype(float)
```

```python
rcd_gen_obra['total_rcd_gen_obra_ton_type'] = rcd_gen_obra['total_rcd_gen_obra_ton'].apply(type)
rcd_gen_obra
```

### Agrupar datos por Autoridad Ambiental

#### Total de RCD generado en obra por Autoridad Ambiental

```python
#pd.set_options('display.max_rows', None)
total_rcd_gen_obra = rcd_gen_obra.groupby(
  'autoridad_ambiental'
  )['total_rcd_gen_obra_ton'].sum().to_frame().reset_index().sort_values(by='total_rcd_gen_obra_ton',
                                                                         ascending=False)
total_rcd_gen_obra
```
